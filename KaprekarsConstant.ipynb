# Kaprekar's constant

```python
import random
```

```python
# Get a random number with 4 digits
num = random.choice(range(1000, 10000))

# Check that all 4 digits are not the same
if len(set(str(num))) == 1:
    num = random.choice(range(1000, 10000))
print(f"Number: {num}")
    
last_diff = 0

# Loop until the difference is 6174
while last_diff != 6174:
    # Sort the digits in ascending and descending order
    num_asc = int(''.join(sorted(str(num))))
    num_desc = int(''.join(sorted(str(num), reverse=True)))
    
    # Calculate the difference
    num_diff = num_desc - num_asc
    
    # Print the number and the difference
    # print(f"Number: {num}\nAsc: {num_asc}\nDesc: {num_desc}\nDifference: {num_diff}")
    print(f"Difference: {num_diff}")
    
    # Update the number and the difference
    num = num_diff
    last_diff = num_diff
```
